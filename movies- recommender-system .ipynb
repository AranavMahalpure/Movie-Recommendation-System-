```python
import numpy as np 
import pandas as pd
```

```python
movies= pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits= pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies=movies[['id','original_title','overview','genres','keywords','original_language','cast','crew','release_date']]
```

```python
movies.head()
```

```python
#removing missing and duplicate data from dataset 
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
# convert genres to main genres
import ast
def convert( obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert1(obj):
    L =[]
    count = 0
    for i in ast.literal_eval(obj):
        if count !=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert1)
```

```python
movies['cast'] =movies['cast'].apply(convert1)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
            
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) 
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x]) 
```

```python
movies.head()
```

```python
movies['tags']=movies['overview'] +movies['genres'] + movies['keywords'] + movies['cast']+ movies['crew']
```

```python
movies.head()
```

```python
new_df =movies[['id','original_title','release_date','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] =  new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] =  new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
 new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
 vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
# movie similar to other movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = new_df[new_df['original_title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list =sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[0:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].original_title)
```

```python
recommend('Avatar')
```
